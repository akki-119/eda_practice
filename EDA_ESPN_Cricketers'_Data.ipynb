```python
#Import pandas as pd
import pandas as pd
```

```python
#read the csv file
df = pd.read_csv("cricket_test_match_data.csv")
df.head()
```

```python
#renaming the columns

df = df.rename(columns = {'NO':'Not_Outs',
                          'HS':'Highest_Inns_Score',
                          'BF':'Balls_Faced',
                          'SR':'Batting_Strike_Rate',
                          'Mat':'Matches'}
              )
```

```python
# Columns to replace "-" with 0
columns_to_replace = ['Balls_Faced', 'Batting_Strike_Rate']

# Replace "-" with 0 in the specified columns
df[columns_to_replace] = df[columns_to_replace].replace('-', 0)
```

```python
df = df.drop_duplicates()
```

```python
# split up span
df['Span'].str.split(pat = '-')
```

```python
df['Rookie_Year'] = df['Span'].str.split(pat='-').str[0]
```

```python
df['Final_Year'] = df['Span'].str.split(pat='-').str[1]
```

```python
#now we can drop span column
df = df.drop(['Span'], axis=1)
```

```python
df.head()
```

```python
# Split up the country from the player
df['Player'].str.split(pat = '(')
```

```python
df['Player_Name'] = df['Player'].str.split(pat='(').str[0]
```

```python
df['Country'] = df['Player'].str.split(pat='(').str[1]
df['Country'] = df['Country'].str.replace(')', '')
```

```python
df = df.drop(['Player'], axis=1)
```

```python
df.head()
```

```python
df.dtypes
```

```python
#For the Highest_Inns_Score column, we have to remove the '*' and convert the dtype to int
#Also, we need to create a new column to show the 'Out' & 'Not Out' status of the batsmen's 'Highest_Inns_Score'

#applying the lambda function
df['Highest_Inns_Score_Not_Out'] = df['Highest_Inns_Score'].apply(lambda x: 'Not Out' if '*' in str(x) else 'Out')

#removing "*" from Highest_Inns_Score
df["Highest_Inns_Score"] = df["Highest_Inns_Score"].str.split(pat='*').str[0]

#convert the dtype to int from str
df["Highest_Inns_Score"] = df["Highest_Inns_Score"].astype('int')
```

```python
df.dtypes
```

```python
df = df.astype({'Rookie_Year':'int', 'Final_Year':'int'})
df.dtypes
```

```python
df.head()
```

```python
#Batting_Strike_Rate should be float

df["Batting_Strike_Rate"] = df["Batting_Strike_Rate"].astype('float')
```

```python
#Matches should be int
df["Matches"] = df["Matches"].astype('int')
```

```python
# we are getting this error because some one of the rows have 138* and due to that "*", it won't change data type to int
df['Matches'] = df['Matches'].str.split(pat="*").str[0]
```

```python
#now we can change it to int
df["Matches"] = df["Matches"].astype('int')
```

```python
#lets's check all the dtypes and do the necessary changes
df.dtypes
```

```python
# balls faced needs to be int
df["Balls_Faced"] = df["Balls_Faced"].astype('int')
```

```python
# now there are some issues with the format of data in Balls_Faced columns
df["Balls_Faced"] = df["Balls_Faced"].str.split(pat="+").str[0]
```

```python
# balls faced needs to be int
df["Balls_Faced"] = df["Balls_Faced"].astype('int')
```

```python
# there's also some NaN values
df["Balls_Faced"] = df["Balls_Faced"].fillna(0)
```

```python
# balls faced needs to be int
df["Balls_Faced"] = df["Balls_Faced"].astype('int')
```

```python
# make a career length column

df["Career_Length"] = df["Final_Year"] - df["Rookie_Year"]
df['Career_Length']
```

```python
df.head()
```

```python
#average career length of the cricketers 
df['Career_Length'].mean()
```

```python
#average Batting_Strike_Rate for cricketers who played over 10 years
df[df['Career_Length'] > 10]['Batting_Strike_Rate'].mean()
```

```python
#number of cricketers who played before 1960
df[df["Rookie_Year"] < 1960]['Player_Name'].count()
```

```python
#Max highest inns score by country
df.groupby('Country')["Highest_Inns_Score"].max().to_frame('Highest_Inns_Country').sort_values('Highest_Inns_Country', ascending=False).reset_index()
```

```python
#Hundreds, Fifties ducks(0) Avg by country
df.groupby('Country')[['100', '50', '0']].mean()
```
